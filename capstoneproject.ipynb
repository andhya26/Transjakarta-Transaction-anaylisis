```python
# Menyiapkan ringkasan informasi
summary_info = {
    'Jumlah Data': len(df),
    'Jumlah Kolom': len(df.columns)
}
summary_info
```

```python
# Membuat ringkasan untuk setiap kolom
summary_df = pd.DataFrame({
    'Tipe Data': df.dtypes,
    'Nilai Null': df.isnull().sum(),
    'Persentase Null (%)': df.isnull().mean() * 100,
    'Jumlah Unik': df.nunique(),
    'Nilai Unik': [df[col].dropna().unique()[:5] for col in df.columns]
})

summary_info, 
summary_df
```

```python
# Mengambil hanya kolom numerik
kolom_numerik = df.select_dtypes(include=[np.number]).columns.tolist()

# Mempersiapkan plot
fig, axes = plt.subplots(len(kolom_numerik), 1, figsize=(15, 4 * len(kolom_numerik)))

# Membuat boxplot untuk setiap kolom numerik
for i, col in enumerate(kolom_numerik):
    axes[i].boxplot(df[col].dropna(), vert=False)
    axes[i].set_title(f'Boxplot - {col}')
    axes[i].set_yticks([1])
    axes[i].set_yticklabels([col])

plt.tight_layout()
plt.show()

```
